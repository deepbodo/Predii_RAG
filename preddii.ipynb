# Data Processing and Filtering for Makes Ford and Toyota

## dropped first column of the txt file that was the line number
## 'RECALL_NUMBER', 'MAKE', 'MODEL', 'YEAR' and the remaining  values in extra column didn;t parse through all the values
## due to this also the condition was iterate through each space in the line so it created column named Eextra:1,Extra:2 etc

```python
import pandas as pd

class RecallDataProcessor:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None

    def load_data(self):
        try:
            self.df = pd.read_csv(self.file_path, sep="\t", header=None, on_bad_lines='skip', engine='python')
        except Exception as e:
            print(f"Error load data: {e}")
        
        return self
    def clean_data(self):
        if self.df is None:
            raise ValueError("Data not load. Need load_data() first.")
        self.df = self.df.drop(columns=[0])
        
        self.df.columns = ['RECALL_NUMBER', 'MAKE', 'MODEL', 'YEAR'] + ['Extra'] * (self.df.shape[1] - 4)
        
        self.df['Extra'] = self.df.iloc[:, 4:].apply(lambda row: ' '.join(row.dropna().astype(str)), axis=1)
        
        self.df = self.df[['RECALL_NUMBER', 'MAKE', 'MODEL', 'YEAR', 'Extra']]

        return self

    def get_processed_data(self):
        if self.df is None:
            raise ValueError("Data not process. Need load_data() and clean_data() first.")
        
        return self.df

def filter_ford_toyota(df):
    filtered_df = df[df['MAKE'].isin(['FORD', 'TOYOTA'])]
    return filtered_df
    
processor = RecallDataProcessor('/kaggle/input/flatrcl/FLAT_RCL.txt')
processor.load_data().clean_data()
df = processor.get_processed_data()

print("Sample of processed data:")
print(df.head())
print(f"\nTotal records: {len(df)}")
filtered_df = filter_ford_toyota(df)
```

# Merged Summary Extraction from 'Extra' Column

In the given approach, I used a simple heuristic to extract a "merged summary" from the `Extra` column of the dataset. 

The key logic used here is:

1. **Hardcoding the 19th Space:**  
   I decided to split the text from the `Extra` column into words and select the content starting from the 19th word onward. This choice was made to quickly identify a segment that likely contains the bulk of the summary information. This is not a professional or dynamic approach, but it helped to save time during data processing by focusing on a pattern based on the dataset structure.

2. **Extracting Text until Full Stop:**  
   Once the words after the 19th space were identified, I took the substring until the first full stop (period) appears. This ensures that the summary text is a coherent sentence or fragment, ending when the thought is completed.

3. **Why This Approach:**  
   The main reason for using this approach was to avoid the complexity of advanced natural language processing techniques like topic modeling, sentence extraction, or deep learning-based summarization. This heuristic allowed for a quick and easy extraction of a portion of text that is most likely to be the "summary" based on the dataset's general structure. 

While this method is far from perfect and lacks adaptability to different text structures, it serves as a simple and fast solution for this specific case.


```python
import re
def extract_summary(text):
    words = text.split()  
    if len(words) >= 19:
        summary_words = words[19:] 
        summary = ' '.join(summary_words)  
        period_pos = next((i for i, word in enumerate(summary_words) if word.endswith('.')), len(summary_words))
        return ' '.join(summary_words[:period_pos + 1])
    return ''
saved_df['Merged_Summary'] = saved_df['Extra'].apply(extract_summary)

print(saved_df[['RECALL_NUMBER', 'MAKE', 'MODEL', 'YEAR', 'Merged_Summary']])
```

# Query Embedding Generation for Retrieval

##The `query_embedding` variable now contains the query's embedding, which is ready for use in a semantic search or comparison with other embeddings.

```python
from sentence_transformers import SentenceTransformer
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
retrieval_model = model1.to(device)

# Generate query embedding
query_embedding = retrieval_model.encode(query, convert_to_tensor=True, device=device)
```

# Getting the Most Similar Documents

In this part, we are finding which documents are most similar to the query based on the embeddings we generated earlier. We use cosine similarity to compare the query with the embeddings in the dataset, and then pick the top `k` most relevant results.

### What We Did:

1. **Set the Limits:**  
   We set two important things:
   - `top_k`: This is how many results we want to see (5 in this case).

2. **Calculate Similarity:**  
   For each document, we calculate how similar it is to our query using **cosine similarity**. We store the similarity score in a new column called `similarity`.

3. **Pick the Best Ones:**  
  Retrieve top-k documents sorted by similarity

### What We Got:
The `retrieved_docs` dataframe now has the top `k` documents that are the most similar to the query. These are the most relevant results based on their embeddings and cosine similarity.


```python
from sentence_transformers import util
top_k = 5

fd['similarity'] = fd['embeddings'].apply(lambda x: util.cos_sim(query_embedding, torch.tensor(x, device=device)).item())

print(fd[['RECALL_NUMBER', 'MAKE', 'MODEL', 'YEAR', 'Merged_Summary', 'similarity']].head(10))
# from sentence_transformers import util
retrieved_docs = fd.nlargest(top_k, 'similarity')


# # number of top results and threshold for similarity
# top_k = 5 
# threshold = 0.1
# # Check the first few rows of similarity scores to inspect them
# # fd['similarity'] = fd['embeddings'].apply(lambda x: util.cos_sim(query_embedding, torch.tensor(x, device=device)).item())
# # print(fd[['RECALL_NUMBER', 'MAKE', 'MODEL', 'YEAR', 'Merged_Summary', 'similarity']].head(10))

# # # Calc similarity between query and embeddings in the 'embeddings' column
# fd['similarity'] = fd['embeddings'].apply(lambda x: util.cos_sim(query_embedding, torch.tensor(x, device=device)).item())
# #get topK
# retrieved_docs = fd[fd['similarity'] > threshold].nlargest(top_k, 'similarity')
# print(retrieved_docs)
```

# Model Loading 

I had limited GPU memory, so I used settings like `low_cpu_mem_usage=True`, `torch_dtype=torch.float16`, and `device_map="auto"` to efficiently load the model on available hardware, automatically using GPU or CPU as needed.


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

base_model = "meta-llama/Llama-3.1-8B"  
tokenizer = AutoTokenizer.from_pretrained(base_model)
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    return_dict=True,
    low_cpu_mem_usage=True,
    torch_dtype=torch.float16,
    device_map="auto", 
    trust_remote_code=True,
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.float16,
    device_map="auto",
)
```
